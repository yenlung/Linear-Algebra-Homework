我們這裡準備看在 $\mathbb{R}^2$ 裡的一根向量, 可以生出什麼樣的東西; 如果是兩根向量呢?

## 標準起手式

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as s
```

## 學畫個向量

畫向量可以用個神秘指令叫 `quiver`。

我們來介紹 `quiver` 最「自然」的用法。

    quiver(X, Y, U, V)
    
記得要畫一個向量當然要告訴 Python 起始點和終點座標, 再記得 `matplotlib` 很愛很愛 x, y 座標分開列, 於是我們有:

* X: 這是我們要畫的起始點 x 座標 (可以很多個)
* Y: 起始點 y 座標
* U: 終點 x 座標
* V: 終點 y 座標

#### one more thing...

如果要畫出我們想像「正常座標」樣, 我們要設幾個參數, 基本上是說用「正常」座標系統, 不要用奇怪單位...

    angles='xy'
    scale_units='xy'
    scale=1
    
我們試試畫兩根向量 (3,2), (-3,1), 起始點都是 (0,0)

```python
X = [0, 0]
Y = [0, 0]
U = [3, -3]
V = [2, -1]
```

```python
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

最後兩行很明顯是要設 xy 座標畫圖範圍, 不設會出亂子。再來我們可能想要設每根向量的顏色, 這很容易做到。

```python
plt.quiver(X, Y, U, V, color=['g', 'b'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

### [練習]

畫出第三根向量, 是前面兩根向量相加的結果。

```python
X = [0, 0, 0]
Y = [0, 0, 0]
U = [3, -3, 0]
V = [2, -1, 1]
plt.quiver(X, Y, U, V, color=['g', 'b', 'r'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

## 起始點其實可以...

像我們的例子中, 三根向量起始點一樣時, 難道不可以說起始點一樣就好? 答案是可以的...

```python
X = [0]
Y = [0]
U = [3, -3, 0]
V = [2, -1, 1]
plt.quiver(X, Y, U, V, color=['g', 'b', 'r'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

Python 有個 `*args`, 基本上就是把 args 中一串東西一一拿出來。所以我們可以耍炫用這樣的方式, 看來更有涵養的感覺。

```python
origin = [0], [0]
U = [3, -3, 0]
V = [2, -1, 1]
plt.quiver(*origin, U, V, color=['g', 'b', 'r'], angles='xy', scale_units='xy', scale=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
```

記得這不是 C 不是 C 不是 C, 學過 C 的同學不要太 high, 以為看到了 C 的指標...
